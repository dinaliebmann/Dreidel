```python
import pandas as pd
import numpy as np

df_dest = pd.read_csv('migrant_flow.csv') # The rows are destination, columns are origin
display(df_dest.head())
display(df_dest.shape)

df_countries = pd.read_csv('country_index.csv')
display(df_countries.head())
display(df_countries.shape)
```

```python
df_dest2017 = df_dest[df_dest['Year']==2017]
df_dest2017 = df_dest2017.drop('Year', axis=1)

df_2017 = pd.melt(df_dest2017, id_vars=['Country'])
df_2017['value'] = df_2017['value'].astype(float)
df_2017 = df_2017[['variable','Country','value']]
df_2017 = df_2017.rename(index=str, columns={'variable': 'Origin',
                                        'Country': 'Destination',
                                        'value': 'Migrants'})
df_2017 = pd.merge(df_2017, df_countries, how='left', left_on='Origin', right_on='Country')
df_2017 = pd.merge(df_2017, df_countries, how='left', left_on='Destination', right_on='Country', suffixes=('_origin','_dest'))
df_2017.head()
```

```python
from floweaver import *

source = 'Country_origin'
target = 'Country_dest'
# waypoint = None
# waypoint2 = 'Continent_dest'

df_2017['source']=source
df_2017['target']=target

original_data = Dataset(df_2017[(df_2017['Country_origin']\
                                .isin(['United Kingdom'])) &
                               (df_2017['Continent_dest']\
                                .isin(['Europe']))])

size = dict(width=1200, height=700,
           margins=dict(left=250, right=250))

source_part = original_data.partition(source)
target_part = original_data.partition(target)
# waypoint_part = original_data.partition(waypoint)
# waypoint2_part = original_data.partition(waypoint2)

nodes = {
    source: ProcessGroup([source], source_part),
    target: ProcessGroup([target], target_part),
#     waypoint: Waypoint(waypoint_part),
#     waypoint2: Waypoint(waypoint2_part),
}

ordering = [
    [source],       # put "origin" on the left...
#     [waypoint],
#     [waypoint2],
    [target],   # ... and "destination" on the right.
]

bundles = [
    Bundle(source, target),#, waypoints=[waypoint2]),   # choose which connections to appear
]

sdd = SankeyDefinition(nodes, bundles, ordering, flow_partition=target_part)#, flow_partition=continents)
weave(sdd, original_data, palette='Set2_8', measures='Migrants').to_widget(**size).auto_save_png('example1.png')
```
